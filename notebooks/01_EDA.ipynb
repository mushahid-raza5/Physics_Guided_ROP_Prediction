# Exploratory Data Analysis (EDA)

This notebook explores the USROP_A 3 N-SH-F-15d drilling dataset:
- Summary statistics of key drilling parameters
- Distribution of Rate of Penetration (ROP)
- Bivariate relationships (WOB, RPM, Torque, TFLO, Depth vs ROP)
- Correlation matrix heatmap


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
import scipy.stats as stats
from statsmodels.stats.diagnostic import het_breuschpagan
from scipy.optimize import curve_fit
```

```python
# Load dataset (skip the units row)
file_path = "../data/USROP_A 3 N-SH-F-15d - Final.csv"
df = pd.read_csv(file_path, skiprows=[1])
print(f"Dataset shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
df.head()
```

```python
cols = ["ROP", "WOB", "RPM", "Torque", "TFLO"]
summary = df[cols].agg(["min", "max", "mean", "std"]).T
print(summary)
```

```python
plt.figure(figsize=(6,4))
df["ROP"].hist(bins=40)
plt.xlabel("ROP (m/h)")
plt.ylabel("Frequency")
plt.title("Histogram of ROP")
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(df["WOB"], df["ROP"], s=5)
plt.xlabel("WOB (kkgf)")
plt.ylabel("ROP (m/h)")
plt.title("WOB vs ROP")
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(df["RPM"], df["ROP"], s=5)
plt.xlabel("RPM")
plt.ylabel("ROP (m/h)")
plt.title("RPM vs ROP")
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(df["Torque"], df["ROP"], s=5)
plt.xlabel("Torque (kN.m)")
plt.ylabel("ROP (m/h)")
plt.title("Torque vs ROP")
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(df["TFLO"], df["ROP"], s=5)
plt.xlabel("Flowrate (L/min)")
plt.ylabel("ROP (m/h)")
plt.title("TFLO vs ROP")
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(df["MD"], df["ROP"], s=5)
plt.xlabel("Measured Depth (m)")
plt.ylabel("ROP (m/h)")
plt.title("Depth vs ROP")
plt.show()
```

```python
corr_vars = ["ROP", "WOB", "RPM", "Torque", "TFLO", "HKLD", "SPPA", "MD"]

corr_matrix = df[corr_vars].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
